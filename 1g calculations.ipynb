```python
import numpy as np
import matplotlib.pyplot as plt

# Σταθερές που προκύπ΄τουν από την κανονικοποίηση. 
a = 9.27972807
b = 4.00221988

# Ορίζουμε τη συνάρτηση
def function(x):
    return (a / 2) * (((x + (b / 2)) / (np.sqrt((x + (b / 2))**2 + b**2))) - ((x - (b / 2)) / (np.sqrt((x - (b / 2))**2 + b**2))))

# Ορίζουμε την παράγωγο της συνάρτησης
def derivative(x):
    h = 1e-5  # Μικρό βήμα για την αριθμητική παραγώγιση
    return (function(x + h) - function(x - h)) / (2 * h)

# Δημιουργούμε τιμές z
x = np.linspace(2.5, 20, 400)

# Υπολογίζουμε τις τιμές y για τη συνάρτηση και την παράγωγό της
y = function(x)
y_prime = derivative(x)

# Σχεδίαση
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Συνάρτηση B(z)')
plt.plot(x, y_prime, label="Παράγωγος B'(z)", linestyle='--')
plt.title('Συνάρτηση μαγνητικού πεδίου και η Παράγωγός της')
plt.xlabel('z')
plt.ylabel('B(z) (kG)')
plt.legend()
plt.grid(True)
plt.show()

```

```python
from scipy.integrate import simps

# Ορισμός του εύρους των τιμών x για το ολοκλήρωμα
x_integral = np.linspace(5, 8.5, 100)
B_integral = function_B(x_integral)  # Υπολογισμός των τιμών B στο διάστημα [2.5, 5]

# Υπολογισμός του ολοκληρώματος της συνάρτησης M(B(x)) για διάφορες τιμές του n
n_values_integral = np.arange(0.5, 2.1, 0.5)  # Τιμές n από 0.5 έως 2 με βήμα 0.5
integral_results = {}

for n in n_values_integral:
    M_Bx = function_M(B_integral, n)
    integral = simps(M_Bx, x_integral)  # Υπολογισμός του ολοκληρώματος με τη μέθοδο Simpson
    integral_results[n] = integral

integral_results
```

```python
# Ορισμός της παράγωγου της συνάρτησης B σε ένα συγκεκριμένο σημείο
def derivative_B(x, h=1e-5):
    return (function_B(x + h) - function_B(x - h)) / (2 * h)

# Υπολογισμός της παράγωγου της συνάρτησης B στο σημείο x 
derivative_at_5 = derivative_B(5)

# Πολλαπλασιασμός των ολοκληρωμάτων της συνάρτησης M(B(x)) με την παράγωγο της συνάρτησης B στο σημείο x
integral_times_derivative = {n: integral * derivative_at_5 for n, integral in integral_results.items()}

derivative_at_5, integral_times_derivative
```

```python
import math

# Παράμετρος d
d = 5

# Υπολογισμός του πολλαπλασιαστή με βάση τη διάμετρο d
multiplier = (10**(-6)) * 79.5775 * math.pi * ((d / 2)**2)

# Πολλαπλασιασμός των προηγούμενων αποτελεσμάτων με τον πολλαπλασιαστή
final_results = {n: value * multiplier for n, value in integral_times_derivative.items()}

multiplier, final_results
```

```python
import sympy as sp

# Όρισμα
x = sp.symbols('x')

# Σταθερές από το φιτάρισμα
a = 9.27972807
b = 4.00221988

# Συνάρτηση
f = (a / 2) * (((x + (b / 2)) / (sp.sqrt((x + (b / 2))**2 + b**2))) - ((x - (b / 2)) / (sp.sqrt((x - (b / 2))**2 + b**2))))

# Παράγωγος
f_prime = sp.diff(f, x)

# ολοκλήρωση
f_integral = sp.integrate(f, x)

f, f_prime, f_integral
```

```python
# Ύψος
x_val = 5

# Παράγωγος στο x 
f_prime_at_5 = f_prime.subs(x, x_val)
f_integral_at_5 = -f_integral.subs(x, x_val)

# Γινόμενο παραγώγου και ολοκληρώματος στο σημείο. 
product_at_5 = f_prime_at_5 * f_integral_at_5
f_prime_at_5, f_integral_at_5, product_at_5
```

```python
import math

d = 3
constant = (10**(-6)) * 79.5775 * math.pi * ((d / 2)**2)

# Γινόμενο και αντικατάσταση στο x
final_result = product_at_5 * constant
final_result
```
