```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    '$q_{k}(mm^{-1})$': [1.570796327, 3.015928947, 4.322831491, 5.365840252, 6.283185307, 6.804689688],
    '$ω_{1}(Hz)$': [16, 7, 5, 10, 47, 40],
    '$ω_{2}(Hz)$': [13, 17, 12, 10, 60, 60],
    '$ω_{3}(Hz)$': [16, 12, 15, 15, 55, 62],
    '$ω_{4}(Hz)$': [31, 33, 25, 40, 55, 85],
    '$ω_{5}(Hz)$': [31, 43, 25, 25, 90, 65],
    '$ω_{6}(Hz)$': [11, 17, 5, 5, 55, 51]
}

df = pd.DataFrame(data)

fig, axs = plt.subplots(6, 1, figsize=(10, 18), sharex=True)

for i, col in enumerate(df.columns[1:], 1):
    axs[i-1].plot(df['$q_{k}(mm^{-1})$'], df[col], marker='*', linestyle='-')
    axs[i-1].set_ylabel(col)
    axs[i-1].grid(True)

axs[-1].set_xlabel('$q_{k}(mm^{-1})$')

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10), sharex=True)

for i, col in enumerate(df.columns[1:], 1):
    ax = axs[(i-1)//3, (i-1)%3]
    ax.scatter(df['$q_{k}(mm^{-1})$'], df[col], s=100, color='orange')
    ax.set_ylabel(col)
    ax.grid(True)

axs[1, 1].set_xlabel('$q_{k}(mm^{-1})$')

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10), sharex=True)

for i, col in enumerate(df.columns[1:], 1):
    ax = axs[(i-1)//3, (i-1)%3]
    ax.scatter(df['$q_{k}(mm^{-1})$'], df[col], s=100, color='orange')
    for j, txt in enumerate(df[col]):
        ax.annotate(f'{txt}', (df['$q_{k}(mm^{-1})$'][j], df[col][j]), textcoords="offset points", xytext=(0,10), ha='center')
    ax.set_ylabel(col)
    ax.grid(True)

axs[1, 1].set_xlabel('$q_{k}(mm^{-1})$')

plt.tight_layout()
plt.show()
```

```python
markers = ['o', 's', '^', 'v', 'd', 'p']

fig, axs = plt.subplots(2, 3, figsize=(15, 10), sharex=True)

for i, col in enumerate(df.columns[1:], 1):
    ax = axs[(i-1)//3, (i-1)%3]
    marker = markers[i-1]
    ax.scatter(df['$q_{k}(mm^{-1})$'], df[col], s=100, color='orange', marker=marker)
    for j, txt in enumerate(df[col]):
        ax.annotate(f'{txt}', (df['$q_{k}(mm^{-1})$'][j], df[col][j]), textcoords="offset points", xytext=(0,10), ha='center')
    ax.set_ylabel(col)
    ax.grid(True)

axs[1, 1].set_xlabel('$q_{k}(mm^{-1})$')

plt.tight_layout()
plt.show()
```

```python
colors = ['orange', 'blue', 'green', 'red', 'purple', 'brown']

fig, axs = plt.subplots(2, 3, figsize=(15, 10), sharex=True)

for i, col in enumerate(df.columns[1:], 1):
    ax = axs[(i-1)//3, (i-1)%3]
    marker = markers[i-1]
    color = colors[i-1]
    ax.scatter(df['$q_{k}(mm^{-1})$'], df[col], s=100, color=color, marker=marker)
    for j, txt in enumerate(df[col]):
        ax.annotate(f'{txt}', (df['$q_{k}(mm^{-1})$'][j], df[col][j]), textcoords="offset points", xytext=(0,10), ha='center')
    ax.set_ylabel(col)
    ax.grid(True)

axs[1, 1].set_xlabel('$q_{k}(mm^{-1})$')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

for i, col in enumerate(df.columns[1:], 1):
    marker = markers[i-1]
    color = colors[i-1]
    ax.scatter(df['$q_{k}(mm^{-1})$'], df[col], s=100, color=color, marker=marker, label=col)
    for j, txt in enumerate(df[col]):
        ax.annotate(f'{txt}', (df['$q_{k}(mm^{-1})$'][j], df[col][j]), textcoords="offset points", xytext=(0,10), ha='center')

ax.set_xlabel('$q_{k}(mm^{-1})$')
ax.set_ylabel('$ω_{k}(Hz)$')
ax.grid(True)
ax.legend(title="Data Sets")

plt.tight_layout()
plt.show()
```
